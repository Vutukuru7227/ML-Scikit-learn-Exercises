```python
X = wine.drop('Cultivator', axis=1)
```

```python
Y = wine["Cultivator"]
```

### * Splitting the dataset with 30% for Test and 70% for Training *

```python
from sklearn.model_selection import train_test_split

X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size = 0.3)
```

### * Scaling the dataset using the StandardScaler *

```python
#Importing the StandardScalar class from the sklearn.preprocessing module
from sklearn.preprocessing import StandardScaler

#Instansiating a StandardScaler object
scaler = StandardScaler()

#Scaling ONLY the training data first
scaler.fit(X_Train)
```

```python
#Scaling the Training and Test data using the StandardScaler
X_Train = scaler.transform(X_Train)
X_Test = scaler.transform(X_Test)
```

### * Creating a Multi Layer Perceptron( Here, two layers with 5 neurons each) using Relu as Activation function *

```python
from sklearn.neural_network import MLPClassifier as MLCP

mlcp = MLCP(hidden_layer_sizes = (5,5) ,activation = 'relu', max_iter = 10000, verbose = True)
```

```python
mlcp.fit(X_Train, Y_Train)
```

### *Using Prediction class to predict the Test data over Training data*

```python
#Predictions

predict = mlcp.predict(X_Test)
```

### * Evaluating the predicted value with the Target *

```python
#Evaluations
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(classification_report(Y_Test, predict))
```

```python
print(confusion_matrix(Y_Test, predict))
```

```python
print(accuracy_score(Y_Test, predict))
```
